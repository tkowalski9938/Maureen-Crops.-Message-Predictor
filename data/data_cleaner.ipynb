```python
import pandas as pd
import re
```

```python
import emoji
emoji.demojize("üòê", delimiters=("", ""))
```

```python
df = pd.read_json("messages.json")
df
# note that each message is fairly short, so very deep rnn
# is probably very useful
```

```python
df = df.drop(labels="date", axis=1)
df
```

```python
df = df[df.user != "Alvan"]
df
```

```python
for index, row in df.iterrows():
    row["content"] = emoji.demojize(row["content"], delimiters=("", ""))
df
```

```python
for index, row in df.iterrows():
    if row["user"] == "Shahithya":
        row["user"] = 0
    elif row["user"] == "Tyler":
        row["user"] = 1
    elif row["user"] == "Josh":
        row["user"] = 2
    elif row["user"] == "Spencer":
        row["user"] = 3
    elif row["user"] == "Bryan":
        row["user"] = 4
    elif row["user"] == "Luka":
        row["user"] = 5
    elif row["user"] == "Jasmine":
        row["user"] = 6
    elif row["user"] == "Maureen":
        row["user"] = 7
    else:
        print("error!")
```

```python
#replaces hyperlinkes with the word, "web_link"
for index, row in df.iterrows():
     row["content"] = re.sub(r"http\S+", "web_link", row["content"])
```

```python
#deletes messages with just weblinks
for index, row in df.iterrows():
    if row["content"] == "web_link":
        df = df.drop([index, index])
df
```

```python
df.to_csv("preprocessed.csv", index=False)
```
